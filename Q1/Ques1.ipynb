# Importing Libraries

```python
import requests
import pandas as pd
import urllib3
from bs4 import BeautifulSoup
from datetime import date, timedelta
import time
```

```python
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)         
```

# Fetching Nifty500 Symbols from Nse

```python
data = pd.read_csv("https://www1.nseindia.com/content/indices/ind_nifty500list.csv") #Reading csv Nifty500 list file hosted on NSE
symbols=data["Symbol"] #Storing All the symbols
```

```python
symbols[402]
```

# Generating DateTime Parameters for iteration

```python
start_date = date(2005, 1, 1)
end_date = date(2022, 7, 8)
delta = timedelta(days=1)
all_dates=[]
while start_date <= end_date:                     #Iterating through all the dates and storing them
    all_dates.append(start_date.strftime("%d%m%Y"))
    start_date += delta
    
```

# Parsing the data

```python
big_data = pd.DataFrame(columns =[2, 5,"date"] ) #Will store all the data
for i in all_dates:
    while True: #True is used so that no date can be skipped, the code keeps re-running till try block gets executed
    
        try:
            url = "https://www1.nseindia.com/archives/equities/mto/MTO_"+i+".DAT"
            html_doc=requests.get(url=url,verify= False) # Sending Requests to the url

            if (html_doc.status_code == 200): #If the url is valid
                data=html_doc.text            #Store the data in form of txt file
                data =data[data.find("Traded Quantity"):] #Remove the strings from txt file

                print(f"Data calculated for all the Nifty500 stocks on "+i[:2]+"-"+i[2:4]+"-"+i[4:] )
                df = pd.DataFrame([x.split(',') for x in data.split('\n')[1:]]) #Transform the txt file to dataframe
                df['date']=i[:2]+"-"+i[2:4]+"-"+i[4:]


                df=df.loc[df[3]=='EQ'] #Only fetching the values for equity series

                df=df[df[2].isin(symbols)] #Storing rows which have the symbols in the symbols list
                df = df.drop(df.columns[[0, 1,3,4,6]], axis=1) #Dropping unnecessary columns

                big_data=big_data.append(df) #Appending the data
        except:
            print("Connection refused by the server..") #In case of Max Queries List, the code sleeps for 5seconds and starts again

            time.sleep(5)
        break

        
```

# Renaming the Big Data File

```python
big_data=big_data.rename(columns={2:"Symbol",5:"Delivery Qty"})
```

```python
big_data
```

# Transforming the data 

```python
dict={} #Dict file is used here due to de fragmantation of data while using pandas.
for i in symbols:
    #converting the big data into symbol-wise data
    
    dict[i+" DATE"] = pd.Series(big_data.loc[big_data['Symbol']==i]['date'].values) 
    dict[i+" DEL QTY"] = pd.Series(big_data.loc[big_data['Symbol']==i]["Delivery Qty"].values)
```

```python
final_dataset=pd.DataFrame(dict)
```
