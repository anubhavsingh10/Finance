# Importing Libraries

```python
from datetime import date
from nsepy import get_history
import pandas as pd
import yfinance as yf
from datetime import date, timedelta
import time
```

# Fetching Nifty500 Symbols from Nse


```python
data = pd.read_csv("https://www1.nseindia.com/content/indices/ind_nifty500list.csv")
data["Symbol"]
symbols=data["Symbol"]
symbols
```

# Gathering the data

```python
dict={}
```

```python
df=pd.DataFrame()
```

```python
for i in symbols:
    while True:
        try:
            print(i)
            stock = get_history(symbol=i,
                           start=date(2005,1,1),
                           end=date(2022,7,8))

            dict[i+" DATE"]=pd.Series(times.strftime('%Y-%m-%d') for times in stock.index)
            dict[i+" OPEN"]=pd.Series(stock["Open"].values)
            dict[i+" HIGH"]=pd.Series(stock["High"].values)
            dict[i+" LOW"]=pd.Series(stock["Low"].values)
            dict[i+" CLOSE"]=pd.Series(stock["Close"].values)
            dict[i+" VOLUME"]=pd.Series(stock["Volume"].values)

        except:
            print("Connection refused by the server..")
            time.sleep(5)
        break
```

```python
df = pd.DataFrame(dict)
```

```python
df.to_excel("output.xlsx")
```
