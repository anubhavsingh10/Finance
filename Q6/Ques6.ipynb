# LTP for NIfty50

```python
ltp=nsepython.nse_quote_ltp("NIFTY")
ltp
```

```python
r_below = ltp*97.5/100
r_above = ltp*102.5/100
```

# Finding the Strike Price for Ques 3 and Ques 4

```python
put_contracts = oi_data[oi_data['Strike Price'].between(r_below, ltp)]
call_contracts = oi_data[oi_data['Strike Price'].between(ltp,r_above)]
```
