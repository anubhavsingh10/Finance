# Importing Libraries

```python
from turtle import pd
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import calendar
import time
```


# Defining Headers and Parametes for the Response

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:102.0) Gecko/20100101 Firefox/102.0',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8',
    'Accept-Language': 'en-US,en;q=0.5',
    # 'Accept-Encoding': 'gzip, deflate, br',
    'Referer': 'https://www.sebi.gov.in/sebiweb/other/OtherAction.do?doMfdTrendsSearch=yes',
    'Origin': 'https://www.sebi.gov.in',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1',
    'Sec-Fetch-Dest': 'document',
    'Sec-Fetch-Mode': 'navigate',
    'Sec-Fetch-Site': 'same-origin',
    'Sec-Fetch-User': '?1',
    'Pragma': 'no-cache',
    'Cache-Control': 'no-cache',
    
}
```

```python
params = {
    'doMfdTrendsSearch': 'yes',
}
```

# Storing Monthly End Dates

```python
iter_dates=[]
for i in range(2010,2023):
    for j in range (1,13):
        if j<=9:
            iter_dates.append(str(calendar.monthrange(i, j)[1])+"-0"+str(j)+"-"+str(i))
        else:
            iter_dates.append(str(calendar.monthrange(i, j)[1])+"-"+str(j)+"-"+str(i))
```

```python
iter_dates=iter_dates[:150]
```

```python
iter_dates
```

```python
final_data = pd.DataFrame(columns =["Trading Date", "Debt/Equity","Gross Purchases(Rs Crore)","Gross Sales(Rs Crore)","Net Investment (Rs Crore)"] )
```

```python
final_data
```

# Retreiving Data

```python
for date in iter_dates:
    while True:
        try:
            data = {
            'currdate': '17-07-2022',
            'loginflag': '0',
            'searchValue': '',
            'date': date,
            'loginEmail': '',
            'loginPassword': '',
            'cap_login': '',
            'moduleNo': '-1',
            'moduleId': '',
            'link': '',
            'yourName': '',
            'friendName': '',
            'friendEmail': '',
            'mailmessage': '',
            'cap_email': '',
            }
            response = requests.post('https://www.sebi.gov.in/sebiweb/other/OtherAction.do', params=params, headers=headers, data=data)  #Sending the requests to the website
            soup = bs(response.content, "html.parser") #Parsing the content using Beautiful Soup Library
            table_rows = soup.find('table', class_="table table-striped table-bordered table-hover background statistics-table") #Fetching the data from table class
            dfs = pd.read_html(str(table_rows)) #Converted all the objects to strings 
            df = pd.concat(dfs) #Appended
            df = df[:-1]
            column_list = df.columns
            df.drop(columns=[column_list[-1]],inplace=True)
            df=df.iloc[:len(df)-2] #Removing Rows containing Total(Monthly) data
            final_data=final_data.append(df)
            print(date)
        except:
            
            print("Connection refused by the server..")
            
            time.sleep(5)
        break
```

# Transforming the Data

```python
final_data=final_data.drop(columns=["Net Investment (Rs Crore)"]) 
```

```python
final_data['Gross Purchases(Rs Crore)'] = final_data['Gross Purchases(Rs Crore)'].astype(float) #Conevrted Str to float
```

```python
final_data['Gross Sales(Rs Crore)'] = final_data['Gross Sales(Rs Crore)'].astype(float)
```

```python
final_data["Net Investment (Rs Crore)"]=final_data['Gross Purchases(Rs Crore)']-final_data['Gross Sales(Rs Crore)']
```

```python
Equity =final_data[final_data['Debt/Equity']=='Equity'] #Divided the data into equity and Debt
Debt  =final_data[final_data['Debt/Equity']=='Debt']
```
